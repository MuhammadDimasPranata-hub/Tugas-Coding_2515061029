**Symmetric and Identity Matrices**

```python
#membuat matriks simetris dengan ukuran 3x3 dan buktikan bahwa itu merupakan matriks simetris
A = [
    [2, 1, 3],
    [1, 4, 5],
    [3, 5, 6]
]
A_T = [[A[j][i] for j in range(3)] for i in range(3)]
is_symmetric = (A == A_T)
is_symmetric
```

```python
import torch      #menggunakan pytorch
I = torch.eye(4)  #membuat matriks identitas I dengan ukuran 4x4
I
```

```python
#membuat vector b_pt dengan ukuran 4x1
b_pt = torch.tensor([
    [1],
    [2],
    [3],
    [4]
], dtype=torch.float32)
torch.matmul(I, b_pt) #kalikan matrix I dengan vector b_pt
```
