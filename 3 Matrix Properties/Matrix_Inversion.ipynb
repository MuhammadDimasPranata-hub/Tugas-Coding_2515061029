**Matrix Inversion**

```python
import numpy as np
```

```python
#membuat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[2, 1],
              [5, 3]])
X
```

```python
#membuat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
Xinv
```

```python
#kalikan X dengan Xinv menggunakan numpy
np.dot(X,Xinv)
```

```python
#Numpy
A = np.array([[4, 2],
              [-5, -3]])
B = np.array([4, -7])
np.linalg.solve(A, B)
```

```python
#Tensorflow
import tensorflow as tf
A = tf.constant([[4.0, 2.0],
                 [-5.0, -3.0]])
B = tf.constant([[4.0],
                 [-7.0]])
tf.linalg.solve(A, B)
```

```python
#Pytorch
import torch
A = torch.tensor([[4.0, 2.0],
                  [-5.0, -3.0]])
B = torch.tensor([[4.0],
                  [-7.0]])
torch.linalg.solve(A, B)
```
