### Frobenius Norm



```python
import numpy as np
```

```python
X = np.array([[1, 2,3], [3, 4,5], [6,7,8]])
X
```

```python
(1**2+2**2+3**2+4**2+5**2+6**2+7**2+8**2)**(1/2)
```

```python
np.linalg.norm(X) # same function as for vector L2 norm
```

# Menggunakan Pytorch

```python
import torch
```

```python
X_pt = torch.tensor([[1, 2,3], [3, 4,5], [6,7,8]]) # torch.norm() supports floats only
X_pt
```

```python
torch.norm(X_pt.float())
```

# Menggunakan Tensorflow

```python
import tensorflow as tf
```

```python
X_tf = tf.Variable([[1, 2,3], [3, 4,5], [6,7,8]]) # tf.norm() also supports floats only
X_tf
```

```python
tf.norm(tf.cast(X_tf, tf.float32))
```
