```python
# 2.lakukan tahapan pada soal 1 dengan menggunakan tensorflow

#TENSORFLOW
import tensorflow as tf

print("\n TENSORFLOW SKALAR")
x_tf = tf.constant(10)
print("Tipe data x_tf:", x_tf.dtype)
```

```python
y_tf = tf.constant(29)
print("Tipe data y_tf:", y_tf.dtype)
```

```python
print("Hasil x_tf + y_tf =", tf.add(x_tf, y_tf).numpy())
```

```python
x_float_tf = tf.constant(2.9)
```

```python
print("Hasil x_float_tf + y_tf =", tf.add(x_float_tf, tf.cast(y_tf, tf.float32)).numpy())
print("Tipe data x_float_tf:", x_float_tf.dtype)
```

```python
print("\nPYTORCH SKALAR")
import torch

x_pt = torch.tensor(10)
print("Tipe data x_pt:", x_pt.dtype)
```

```python
y_pt = torch.tensor(29)
print("Tipe data y_pt:", y_pt.dtype)
```

```python
print("Hasil x_pt + y_pt =", (x_pt + y_pt).item())
```

```python
x_float_pt = torch.tensor(5.5)
print("Hasil x_float_pt + y_pt =", (x_float_pt + y_pt).item())
print("Tipe data x_float_pt:", x_float_pt.dtype)
```
