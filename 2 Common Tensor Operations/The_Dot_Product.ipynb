```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# membuat vector x dan y ukuran 1x3
x = np.array([1, 2, 3])   # vector x
y = np.array([4, 5, 6])   # vector y
```

```python
# membuat vector x_pt dan y_pt ukuran 1x3
x_pt = torch.tensor([1, 2, 3])  # vector x pytorch
y_pt = torch.tensor([4, 5, 6])  # vector y pytorch
```

```python
# membuat vector x_tf dan y_tf ukuran 1x3
x_tf = tf.constant([1, 2, 3])  # vector x tensorflow
y_tf = tf.constant([4, 5, 6])  # vector y tensorflow
```

```python
# menghitung dot product antara vector x dan vector y
dot_numpy = np.dot(x, y)  # perkalian dot product

print("Dot Product (NumPy):", dot_numpy)
```

```python
# menghitung dot product antara vector x_pt dan vector y_pt
dot_torch = torch.dot(x_pt, y_pt)  # dot product pytorch

print("Dot Product (PyTorch):", dot_torch.item())
```

```python
# menghitung dot product antara vector x_tf dan vector y_tf
dot_tf = tf.tensordot(x_tf, y_tf, axes=1)  # dot product tensorflow

print("Dot Product (TensorFlow):", dot_tf.numpy())
```

```python
#lakukan perhitungan excercise menggunakan collab
# 1. Transpose matriks Y
Y = np.array([[42, 4, 7, 99],
              [-99, -3, 17, 22]])  # matriks Y

Y_T = Y.T  # transpose matriks

print("\nTranspose Y:\n", Y_T)
```

```python
# 2. Hadamard product (perkalian elemen per elemen)
A = np.array([[25, 10],
              [-2, 1]])  # matriks A

B = np.array([[-1, 7],
              [10, 8]])  # matriks B

hadamard = A * B  # perkalian elemen per elemen

print("\nHadamard Product:\n", hadamard)
```

```python
# 3. Dot product w dan x
w = np.array([-1, 2, -2])  # vector w
x2 = np.array([5, 10, 0])  # vector x

dot_wx = np.dot(w, x2)  # dot product

print("\nDot Product w dan x:", dot_wx)
```
